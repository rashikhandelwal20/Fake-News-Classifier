```python
df_final['text']= df_final['text'].apply(lambda x: x.lower())
```

```python
def remove_special_chars(text):
    
    result = ""
    
    for i in text:
        if i.isalnum() or i == ' ':
            result  = result + i
        else:
            result = result + ""
    return result.strip()
```

```python
remove_special_chars('subject: want to accept credit cards ? 126432211  aredit cpproved  no cecks  do it now  126432211')
```

```python
df_final['text']=df_final['text'].apply(remove_special_chars)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(stop_words='english', max_features=10000)
```

```python
X= cv.fit_transform(df_final['text']).toarray()
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y, train_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred= clf.predict(X_test)
```
